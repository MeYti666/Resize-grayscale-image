```python
from PIL import Image
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
first_img = Image.open("img.jpg")
print("------------------------")
print(first_img.size)
print("------------------------")
first_img
```

```python
gray_img = first_img.convert("L")
```

```python
gray_img.save("gray_img.jpg")
gray_img
```

```python
img_size = mpimg.imread('gray_img.jpg')
img_size.shape
```

```python
from PIL import Image
```

```python
og_img = Image.open('gray_img.jpg')
og_img.size
```

```python
img_1024 = og_img.resize((1024, 1024))
img_1024.save("1024_img.jpg")
img_1024.size
```

```python
img_512 = og_img.resize((512, 512))
img_512.save("512_img.jpg")
img_512.size
```

```python
img_32 = og_img.resize((32, 32))
img_32.save("32_img.jpg")
img_32.size
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(img_1024, cmap=plt.get_cmap('gray'))
axs[0].set_title('Image 1024x1024')

axs[1].imshow(img_512, cmap=plt.get_cmap('gray'))
axs[1].set_title('Image 512x512')

axs[2].imshow(img_32, cmap=plt.get_cmap('gray'))
axs[2].set_title('Image 32x32')

plt.tight_layout()
plt.show()
```
